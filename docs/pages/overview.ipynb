````python
from numpy.random import random
import matplotlib.pyplot as plt
````

## Plot theme

Vanilla plot:

````python

plt.scatter(random(10), random(10)) ;
````

Style a single plot:

````python
from vsnb import vscode_theme as theme

with theme():
    plt.scatter(random(10), random(10)) ;
````

Default plotting style is preserved:

````python
plt.scatter(random(10), random(10)) ;
````

Set theme for all plots:

````python
import vsnb
vsnb.set_vscode_theme()

plt.scatter(random(10), random(10)) ;
````

````python
plt.scatter(random(10), random(10)) ;
````

Even with global styling set, the context manager is useful for setting figsize and font_scale for individual plots:

````python
with theme(font_scale=1.5, figsize=(2,1)):
    plt.scatter(random(10), random(10)) ;
````

````python
with theme(font_scale=0.8, figsize=(7,3)):
    plt.subplot(121).scatter(random(10), random(10))
    plt.subplot(122).scatter(random(10), random(10))
````
